## **Centroides**

Los centroides pueden ser de utilidad en análisis posteriores, por tal razón, es una buena idea manipularlos y evidenciar como se usan a través de geopandas.

```python
med_bar_shp['centroide'] = med_bar_shp.centroid
```

```python
med_bar_shp
```

```python
med_bar_shp.iloc[0].centroide
```

```python
med_bar_shp["area"] = med_bar_shp.area
```

Puedo calcular la distancia entre puntos usando el metodo **distance**. Graficamente los centroides se pueden ver a continuación

```python
axis = med_bar_shp["geometry"].plot(figsize=(20,12))
med_bar_shp["centroide"].plot(ax=axis, color = "red")
```

## **Generación de nuevas geometrías**



Un ubjeto importante en el análisis geospacial son los buffers, los cuales son una herramienta utilizada para crear una zona alrededor de un objeto geográfico, como un punto, una línea o un polígono. El buffer se define mediante una distancia específica, que puede ser medida en unidades de longitud, como metros o kilómetros. Con geopandas podemos definir un buffer de la siguiente manera

```python
med_bar_shp["buffer"] = med_bar_shp.buffer(10000)
```

```python
med_bar_shp["buffer_centroide"] = med_bar_shp["centroide"].buffer(100)
```

```python
med_bar_shp[med_bar_shp['COMUNA']=='07'].plot()
```

## **Georreferenciación de los casos de hurtos**

A continuación leemos los datos asociados a hurtos a personas en la ciudad de Medellín durante el 2018-2022

```python
df = pd.read_excel('/content/drive/MyDrive/Maestria_eafit/Hurto.xlsx')
```

```python
df.shape
```

Filtramos rápidamente aquellas coordenadas fuera del rango de la ciudad

```python
df =df[df['Geo X Hurto'] > -76]
```

```python
df.shape
```

```python
df[['Geo X Hurto','Geo Y Hurto']]
```

A continuación ponemos las coordenadas como valores geometricos

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df['Geo X Hurto'], df['Geo Y Hurto'])
)
```

```python
gdf.geometry
```

Vemos inciialmente como se distribuyen los hurtos en el espacio, esto sin tener en cuenta la capa geometríca de la ciudad.

```python
gdf.plot()
```

De la siguiente manera tenemos los hurtos ubicados en el mapa de Medellín

```python
axis = med_bar_shp.plot(figsize=(20,14))
gdf.plot(ax=axis, color = "red",markersize =0.1)
```
